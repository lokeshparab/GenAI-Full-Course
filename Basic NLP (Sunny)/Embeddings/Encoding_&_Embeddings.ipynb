# Import Data

```python
import pandas as pd
data = pd.DataFrame({
    "text":["people watch cricket","cricket watch cricket","people give comment","cricket give comment"],
    "output":[1,1,0,0]
})
print(data.columns)
data
```

# BOW

Word dimensions are arrangement in **Alphabetical sequence**

```python
from sklearn.feature_extraction.text import CountVectorizer

BOW = CountVectorizer()

bow_document_matrix=BOW.fit_transform(data["text"])

feature_names = BOW.get_feature_names_out()

print(feature_names)
print(BOW.vocabulary_)
data
```

```python
bow_document_matrix.toarray()
```

# N-gram (Bi-gram)

The n-gram are token in alphabetical sequence

```python
n = 2
bigram = CountVectorizer(ngram_range=(n,n))

bigram_document_matrix=bigram.fit_transform(data["text"])

feature_names = bigram.get_feature_names_out()

print(feature_names)
print(bigram.vocabulary_)
data
```

```python
bigram_document_matrix.toarray()
```

```python
# mixture of 1 gram and 2 gram

mix = CountVectorizer(ngram_range=(1,2))

mix_document_matrix=mix.fit_transform(data["text"])

feature_names = mix.get_feature_names_out()

print(feature_names)
print(mix.vocabulary_)
data
```

```python
mix_document_matrix.toarray()
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

tfidf_vocal=tfidf.fit_transform(data["text"])

feature_names = tfidf.get_feature_names_out()

print(feature_names)

print(tfidf.vocabulary_)
print(tfidf.idf_)
```

```python
tfidf_vocal.toarray()
```

## Checking Similarity score between 2 words

```python
print("Similarity in men and women",model.similarity("man","woman"))
print("Similarity in sunny and rich",model.similarity("sunny","rich"))
print("Similarity in java and python",model.similarity("java","python"))
print("Similarity in java and javascript",model.similarity("java","javascript"))
```

## Checking the match from the list

```python
print(model.doesnt_match(["PHP","JAVA","PERL","C","SNAKE"])) # this giving mischeif result
print(model.doesnt_match(["Human","Bird","Animal","Cat"]))
```
